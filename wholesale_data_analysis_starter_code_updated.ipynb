## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['qty'] * df['unit_price']
totals = df[['subtotal','qty', 'unit_price']].head()
totals
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
###	unit_price	unit_weight	qty	total_weight	shipping_price
### 0 	1096.80	7.50	105	787.50	5512.50

df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = np.where(df['total_weight'] > 50, df['total_weight'] * 7, df['total_weight'] * 10)
df[['unit_price','unit_weight','total_weight','shipping_price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
### 	line_subtotal	shipping_price	line_price
### 0	115164.00	5512.50	131839.08
### 1	523.95	312.90	914.26
### 2	527.28	458.64	1077.12

df['total_price'] = round(((df['subtotal'] + df['shipping_price']) *1.0925),2)
df['tax_payable'] = round(((df['subtotal'] + df['shipping_price']) *0.0925),2)
df[['subtotal', 'shipping_price','tax_payable','total_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df[['unit_cost', 'qty', 'shipping_price','line_cost']].head()
#df
```

```python
# Create a column for the profit of each line using line cost and line price
# 	total_price	line_cost	line_profit
# 0	131839.08	85597.05	46242.03
# 1	914.26	629.79	284.47
df['line_profit'] = df['total_price'] - df['line_cost']
round(df[['total_price', 'line_cost','line_profit']].head(),2)
```

Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

top_five_client_ids = [33615, 66037, 46820, 38378, 24741] ## 

def clients_total(client_id, column):
    client_df = df.loc[df['client_id'] == client_id, 'total_price']
    return round(client_df.sum(),2)

for client_id in top_five_client_ids:
    print(f"{client_id} total: ${clients_total(client_id, 'total_price')}")
```

```python

def clients_total(client_id, column):
    client_df = df.loc[df['client_id'] == client_id, column] # where total price == column
    return round(client_df.sum(),2)

for client_id in top_five_client_ids:
    print(f"{client_id} total: ${clients_total(client_id, 'total_price')}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

data_dictionary = {'client_id': top_five_client_ids }
columns = ['qty','shipping_price','total_price','line_cost','line_profit'] 
for col in columns:
    data_dictionary[col] = [clients_total(client_id, col) 
                           for client_id in top_five_client_ids]
    
new_data_frame = pd.DataFrame(data_dictionary)
new_data_frame
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 
# Rename the columns to reflect the change in the money format. 
```

```python
# Define the money columns. 

#data_dictionary = {'client_id': top_five_client_ids }
columns = ['line_profit','total_price', 'shipping_price','line_cost'] 

def currency_million(amount):
    return amount/1000000

for col in columns:

    new_data_frame[col] = new_data_frame[col].apply(currency_million)
new_data_frame
```

```python
# Format the data and rename the columns to names suitable for presentation.

new_data_frame = new_data_frame.rename(columns={'client_id': "Client Id",
                               'qty': "Qauntity Purchased",
                               'shipping_price': "Shipping Costs" ,
                               'total_price': "Gross Revenue",
                               'line_cost': "Total Costs",
                               'line_profit':"Total Profit"}
                      )
new_data_frame
```

```python
# Sort the updated data by "Total Profit (millions)" from highest to lowest and assign the sort to a new DatFrame.

sort_new_data = new_data_frame.sort_values("Total Profit", ascending=False)
sort_new_data
```
